## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, 
# standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following 
# properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

# Call the DataFrame, copy df
clean_df = clean_df.copy()   
```

```python
# Groupby Drum Regimen 
grouped_clean_df = clean_df.groupby('Drug Regimen')
grouped_clean_df.head()
```

```python
# mean tumor volume of each drug regimen
tumor_volume_mean = grouped_clean_df.mean()['Tumor Volume (mm3)']
tumor_volume_mean
```

```python
# median tumor volume of each drug regimen
tumor_volume_median = grouped_clean_df.median()['Tumor Volume (mm3)']
tumor_volume_median
```

```python
# variance of tumor volume for each drug regimen 
tumor_volume_variance = grouped_clean_df.var()['Tumor Volume (mm3)']
tumor_volume_variance
```

```python
# standard deviation of tumor volume of each drug regimen
tumor_volume_sd = grouped_clean_df['Tumor Volume (mm3)'].std()
tumor_volume_sd
```

```python
# standard error of the mean (SEM) of the tumor volume of each drug regimen
tumor_volume_sems = grouped_clean_df['Tumor Volume (mm3)'].sem()
tumor_volume_sems
```

```python
# Assemble the resulting series into a single summary dataframe.
summary_df = pd.DataFrame({
     'Mean': tumor_volume_mean, 
     'Median': tumor_volume_median,
     'Variance':tumor_volume_variance,
     'SD': tumor_volume_sd,
     'SEM': tumor_volume_sems
    })
```

```python
# Generate a summary statistics table of mean, median, variance, 
# standard deviation, and SEM of the tumor volume for each regimen
summary_df.head(10)
```

```python
# Using the aggregation method, produce the same summary statistics 
# in a single line
aggregated_single_line = grouped_clean_df.agg(['mean', 'median','var','std', 'sem'])['Tumor Volume (mm3)']
aggregated_single_line
```

## Quartiles, Outliers and Boxplots

```python
#restated df for myself 
clean_df.head()
```

```python
#list unique id
ids = clean_df['Mouse ID'].unique().tolist()
ids[0]
```

```python
# Calculate the final tumor volume of each mouse across four of the 
# treatment regimens: Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
ids = list()
max_vals = list()
for mouse_id in clean_df['Mouse ID'].unique():
    ids.append(mouse_id)
    mask = clean_df['Mouse ID']==mouse_id
    max_vals.append(clean_df[mask].Timepoint.max())

vals_df = pd.DataFrame({
    'Mouse ID': ids,
    'Timepoint': max_vals
})
```

```python
# Merge this group df with the original dataframe to get the tumor volume 
# at the last timepoint
merged_df = pd.merge(clean_df, vals_df,on='Mouse ID',how='outer')
merged_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
final_volume = []

# Calculate the IQR and quantitatively determine if there are any 
# potential outliers. 
for drugs in treatments: 
    quartiles = merged_df[merged_df['Drug Regimen']==drugs]['Tumor Volume (mm3)'].quantile([.25,.5,.75])
    lower_quartile = quartiles[.25]
    upper_quartile = quartiles[.75]
    iqr = upper_quartile - lower_quartile
    
    # Locate the rows which contain mice on each drug and get the 
    # tumor volumes
    mice = merged_df['Drug Regimen']==drugs,['Tumor Volume (mm3)']
    
    # add subset 
    final_volume.append(mice)
    
    # Determine outliers using upper and lower bounds
    lower_bound = lower_quartile - (1.5*iqr)
    upper_bound = upper_quartile + (1.5*iqr)
    outlier_tv = merged_df[(merged_df['Tumor Volume (mm3)'] < lower_bound) | (merged_df['Tumor Volume (mm3)'] > upper_bound)]['Tumor Volume (mm3)']

print(lower_bound)
print(upper_bound)
print(outlier_tv)
```

```python
merged_df = merged_df.rename(columns={'Timepoint_x':'Timepoint_varied','Timepoint_y':'Timepoint_final'})
merged_df
```

```python
# create df with final tumor volume of each mouse across 
# four regimens of interest
final_tumor_volume_df = merged_df.loc[merged_df['Timepoint_varied']==merged_df['Timepoint_final']]
final_tumor_volume_df.head()
```

```python
#group by 4 drug regimens
capodf = final_tumor_volume_df.loc[final_tumor_volume_df['Drug Regimen']=='Capomulin'].groupby('Drug Regimen')
ramdf = final_tumor_volume_df.loc[final_tumor_volume_df['Drug Regimen']=='Ramicane'].groupby('Drug Regimen')
infubdf = final_tumor_volume_df.loc[final_tumor_volume_df['Drug Regimen']=='Infubinol'].groupby('Drug Regimen')
ceftdf = final_tumor_volume_df.loc[final_tumor_volume_df['Drug Regimen']=='Ceftamin'].groupby('Drug Regimen')
```

```python
# create df for data we want: drug regimens 'Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin' 
final_tumor_volume_cleaned_df = (list(capodf['Tumor Volume (mm3)'])[0][1],
                                 list(ramdf['Tumor Volume (mm3)'])[0][1],
                                 list(infubdf['Tumor Volume (mm3)'])[0][1],
                                 list(ceftdf['Tumor Volume (mm3)'])[0][1])
final_tumor_volume_cleaned_df
```

```python
#plot box plot 
fig1,ax,=plt.subplots()
ax.boxplot(final_tumor_volume_cleaned_df)
plt.xticks([1, 2, 3, 4], ["Capomulin", "Ramicane","Infubinol","Ceftamin"])
plt.title('Tumor Volume (mm3) of Four Drug Regimens', fontsize=10)
plt.ylabel('Tumor Volume (mm3)', fontsize=10)
plt.xlabel('Drug Regimens', fontsize=10)
plt.grid(axis='y', alpha=.5, linewidth=1.5)
```

## Correlation and Regression

```python
# Calculate the correlation coefficient for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(x, y)
print(f'The correlation between the mouse weight and average tumor volume for mice treated with Capomulin is {round(correlation[0],2)}')
```

```python
# Code linear regression
correlation = st.pearsonr(x, y)
(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
regression = x * slope + intercept
line_eq = "y =" + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x, y)
plt.plot(x, regression, "r-")
plt.annotate(line_eq, (42,16), fontsize = 10, color = "blue")
```

```python
# Add labels to scatterplot with linear regression 
plt.xlabel('Tumor Voulme (mm3)')
plt.ylabel('Weight (g)')
plt.title('Capomulin Regimen: Average Tumor Volume v. Mouse Weight')
plt.show()
```
